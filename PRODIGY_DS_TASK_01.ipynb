```python
df_years = df[['Country Name'] + [str(year) for year in range(1960, 2024)]]
df_years = df_years.set_index('Country Name')
df_years = df_years.transpose()
df_years = df_years.astype(float)

# Plotting a line plot for population trends of a few countries
countries_to_plot = ['India', 'United States', 'China']
df_years[countries_to_plot].plot(figsize=(12, 6))
plt.title("Population Trends (1960-2023)")
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(countries_to_plot)
plt.show()
```

```python
# Ensure that only numeric columns are included, and convert all columns to numeric, coercing errors.
df_years = df[['Country Name'] + [str(year) for year in range(1960, 2024)]]
df_years = df_years.set_index('Country Name')

# Convert all columns to numeric, forcing errors to NaN
df_years = df_years.apply(pd.to_numeric, errors='coerce')

# Now apply cumulative sum
df_cumulative = df_years.cumsum(axis=1)

# Plot cumulative population growth for India as an example
df_cumulative.loc['India'].plot(figsize=(12, 6))
plt.title("Cumulative Population Growth for India (1960-2023)")
plt.xlabel('Year')
plt.ylabel('Cumulative Population')
plt.show()
```

```python
# Calculate global population by summing across countries for each year
global_population = df.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']).sum()

# Plot the global population trend
global_population.plot(figsize=(12, 6))
plt.title("Global Population Growth (1960-2023)")
plt.xlabel("Year")
plt.ylabel("Global Population")
plt.show()
```
